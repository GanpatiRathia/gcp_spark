```python
data = range(1, 100) # create a list having numbers form 1 to 99
print(list(data)) # printing the list
```

```python
rangeRDD = sc.parallelize(data, 8) # creating RDD with 8 partition having numbers present in the previously created list
rangeRDD.collect() # print data stored in RDD
```

```python
type (rangeRDD) # prints type of RDD created
```

```python
rangeRDD.getNumPartitions() #prints number of partitions
```

```python
def sub(value): # method to apply transformation
    return (value-1)
sub(10)
```

```python
newRDD = rangeRDD.map(sub) # applying sub method to each data item present in the cluster
newRDD.collect() 
```

```python
rangeRDD.collect() # no change immutable
```

```python
def filter10(val):
    if val > 10:
        return False
    else:
        return True
```

```python
newFillRDD = rangeRDD.filter(filter10) # applying filter to rangeRDD
newFillRDD.collect()
```

```python
newFillRDD1 = rangeRDD.filter(lambda x:x < 10)
newFillRDD.collect()
```

```python
newFillRDD.first(); # return first element of RDD.
newFillRDD.take(100)
```
