### remove the first column and show the output
### show the data corresponding to 'Iris-setosa'

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
  .appName('3.1. Google Cloud Storage (CSV) & Spark DataFrames') \
  .getOrCreate()
```

```python
path = 'gs://iriscsvfiles/iris.csv'
```

```python
df1 = spark \
  .read \
  .option ( "inferSchema" , "true" ) \
  .option ( "header" , "true" ) \
  .csv (path)

df1.printSchema()
```

```python
df1
```

```python
df1.show()
```

```python
df2 = df1.drop("Id")

# Show the dataset
df2.show()
```

```python
# Filter the data for "iris-setosa" class
iris_setosa_data = df2.filter(df2["Species"] == "Iris-setosa")
iris_setosa_data.show()
```
